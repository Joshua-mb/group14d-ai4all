```python
import pandas as pd
import numpy as np

stocks_list = ['AAPL', 'MSFT', 'GOOG', 'META', 'INTC']

for stock in stocks_list:
    file_location = f'./{stock} Data/{stock}_finance_data.csv'
    df = pd.read_csv(file_location)
    print(f'Processing {stock} data')
    df['price_change'] = df['Close'] - df['Close'].shift(1)
    df['target'] = np.sign(df['price_change']).astype('Int64')
    output_location = f'./{stock} Data/{stock}_finance_data.csv'
    df.to_csv(output_location, index=False)
    
```
