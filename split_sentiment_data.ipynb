## 1. Retrieve sentiment data for all stocks

```python
import pandas as pd

sentiment_df = pd.read_csv('stock_tweets_with_sentiment.csv')
sentiment_df.head()
```

## 2. Select only tech stocks

```python
# Select only specific tech stocks from the dataset
tech_sentiment_df = sentiment_df.loc[sentiment_df['Stock Name'].isin(['AAPL', 'GOOG', 'INTC', 'META', 'MSFT'])]
tech_sentiment_df
```

## 3. Group data by stock name and create CSV file

```python
# Creating dataframes for each stock and converting into CSV files
stock_groups = tech_sentiment_df.groupby('Stock Name')
stock_dfs = {stock: group.copy() for stock, group in stock_groups}

for stock_name in stock_dfs:
    curr_df = stock_dfs[stock_name]
    curr_df_sorted = curr_df.sort_values(by='Date')
    curr_df_sorted.to_csv(stock_name + "_sentiment_data.csv", index = False)
```
